```python
import pandas as pd
```

```python
data=pd.read_csv('flight.csv')
data
```

```python
l1=[]
l2=[]
for i in range(len(data)):
    temp=data['f_time_from'][i].split(' ')
    l1.append(temp[0])
    l2.append(temp[1])
data=data.drop(['f_time_from'], axis=1)

data.insert(4,'f_time_from',l1)
data.insert(5,'f_date_from',l2)
l1=[]
l2=[]
for i in range(len(data)):
    temp=data['f_time_to'][i].split(' ')
    l1.append(temp[0])
    l2.append(temp[1])
data=data.drop(['f_time_to'], axis=1)

data.insert(4,'f_time_to',l1)
data.insert(5,'f_date_to',l2)
data
```

```python
for i in range(len(data['f_date_to'])):
   tmp=data['f_date_to'][i].split('/')
   if tmp[1]!='05':tmp=tmp[0]+'/01/'+tmp[2]
   else:tmp=tmp[0]+'/02/'+tmp[2]
   data['f_date_to'][i]=tmp

    
```

```python
data
```

```python
for i in range(len(data['f_date_from'])):
   data['f_date_from'][i]=data['f_date_from'][i].replace('2021','2023')
```

```python
data.to_json('flights.json',orient='records')
```
